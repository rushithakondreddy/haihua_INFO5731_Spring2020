```
import nltk
import pandas as pd
df = pd.read_csv('review_data.csv')
df['word_count'] = df['review_content'].apply(lambda x:len(str(x).split(" ")))
df
```

```
df['review_content'] = df['review_content'].str.replace('[^\w\s]','')  #removal of punctuation
df['review_content'] = df['review_content'].apply(lambda x: " ".join(x.lower() for x in x.split()))#lower case
from nltk.corpus import stopwords #removal of stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
df['review_content'] = df['review_content'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```
#converting to a single line file
line = ''
for i in range(0,len(df['review_content'])):
  line+= df['review_content'][i]
print(line)
```

```
#trigrams
nltk.download('punkt')
words = nltk.word_tokenize(line)
result_trigrams = nltk.trigrams(words)
freq_trigrams = nltk.FreqDist(result_trigrams)
freq_trigrams
```

```
#bigrams
result_bigrams = nltk.bigrams(words)
import re
from itertools import tee, islice
import collections
def ngrams(lst, n):
  tlst = lst
  while True:
    a, b = tee(tlst)
    l = tuple(islice(a, n))
    if len(l) == n:
      yield l
      next(b)
      tlst = b
    else:
      break
line2 = collections.Counter(ngrams(line.split(), 2))
print(line2)

```

```
#join bigrams
s1 = ''
for w1,w2 in result_bigrams:
  s1 = s1 + ' ' + w1 + ' ' + w2
list = s1.split()
print(list)

```

```
for w1,w2 in line2:      #probability
  den = list.count(w1)
  num = line2[(w1,w2)]
  if den == 0:
    break
  prob = num/den
  print(w1,w2,prob)
```
